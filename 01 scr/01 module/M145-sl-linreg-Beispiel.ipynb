```
import numpy as np

# Beispiel: Liste von x- und y-Werten
x = np.array([49,  49,  51, 56, 56,  58, 58, 59, 59, 60, 64, 65, 67, 67, 68, 70, 70, 73, 74, 76, 79, 80, 82, 82, 85, 86, 88, 97, 100, 103])
y = np.array([156, 157, 161, 170, 164, 159, 161, 160, 166, 168, 166, 165, 171, 168, 168, 185, 160, 188, 179, 180, 181, 179, 194, 190, 180, 182, 186, 197, 194, 206])

# Schätzen von a und b mit der Methode der kleinsten Quadrate
A = np.vstack([x, np.ones(len(x))]).T
a, b = np.linalg.lstsq(A, y, rcond=None)[0]

print(f"Regressionsfunktion: y = {a:.4f} x + {b:.4f}")
```

Dieser Code stapelt vertikal zwei Arrays, x und ein Array aus Einsen mit derselben Länge wie x, und erstellt ein neues 2D-Array

```
A.shape
```

Zeile 9:

Dieser Code berechnet die Lösung der kleinsten Quadrate für eine lineare Gleichung Ax=y, wobei A eine 2D-Matrix ist, die durch das Array A repräsentiert wird, x und y Vektoren sind, die durch a bzw. b repräsentiert werden. Die Funktion np.linalg.lstsq() gibt eine Folge von Ausgaben zurück; die erste Ausgabe enthält die Lösung der linearen Gleichung. Daher wird [0] verwendet, um diese Lösung zu extrahieren, und weist a bzw. b das erste und zweite Element dieser Lösung zu.

```
a, b
```
