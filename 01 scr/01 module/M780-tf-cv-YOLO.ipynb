Die Objekterkennung ist eine Computertechnologie im Zusammenhang mit Computer Vision und Bildverarbeitung, die sich mit der Erkennung von Instanzen semantischer Objekte einer bestimmten Klasse (z. B. Menschen, Gebäude oder Autos) in digitalen Bildern und Videos befasst. Zu den gut erforschten Domänen der Objekterkennung gehören die Gesichtserkennung und die Fußgängererkennung. Die Objekterkennung hat Anwendungen in vielen Bereichen der Computer Vision, einschließlich Bildabruf und Videoüberwachung.








**YOLO**
steht für "You Only Look Once" und ist ein Framework für Objekterkennung in Echtzeit. Die aktuellste **offizielle** Version von YOLO ist YOLOv5, die im Juni 2020 veröffentlicht wurde. Es ist möglich, dass es individuelle Entwicklungen von YOLO gibt, die nicht offiziell veröffentlicht wurden und als YOLOv8 bezeichnet werden, aber ohne nähere Informationen kann ich dazu keine genaue Auskunft geben.
