# Best of

| Operation                  | Pandas                                         |   |
|----------------------------|------------------------------------------------|---|
| Read CSV                   | pd.read_csv(file)                              |   |
| Print first 10 (or k) rows | df. head(10)                                   |   |
| Dimensions                 | df.shape                                       |   |
| Datatype                   | df.dtyes                                       |   |
| Filter Data                | df[df. cname> 10]                              |   |
| Select column(s)           | df.cname                                       |   |
| Sort                       | df.sort_values("cname")                        |   |
| Fill NaN                   | df.cname. fillna(O)                            |   |
| Join                       | pd.merge(dfl, df2, on="cname", how="inner")    |   |
| Concatenate                | pd.concat((dfl, df2))                          |   |
| Group                      | df. groupby("cname").agg_col. mean()/.sum()/  |   |
| Unique values              | df.cname.unique()                              |   |
| Rename column              | df.rename(columns = {"old_name": "new_name"})  |   |
| Delete column              | df .drop(columns =  ["cname"])                 |   |
| Delete row                 | df = df.drop(df[df.cname = condition].index)
